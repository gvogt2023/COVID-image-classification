```python
import os
import cv2
import matplotlib.pyplot as plt

# Define the path to the directory containing the images
dir_path = "COVID-19 Dataset/CT/COVID/"

# Loop through the first 10 images in the directory and plot them
fig, axs = plt.subplots(2, 5, figsize=(10, 4))
axs = axs.ravel()

for i, filename in enumerate(os.listdir(dir_path)[:10]):
    img_path = os.path.join(dir_path, filename)
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    axs[i].imshow(img, cmap='gray')
    axs[i].set_title(filename)
    axs[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Path to the directory containing the X-ray images
dir_path = "COVID-19 Dataset/X-ray/COVID/"

# Initialize a list to store pixel values of all images
pixel_values = []


# Loop through the first 100 images in the directory in batches of 8 and append pixel values to the list
batch_size = 8
for i in range(0, 100, batch_size):
    batch = os.listdir(dir_path)[i:i+batch_size]
    for filename in batch:
        img_path = os.path.join(dir_path, filename)
        img = cv2.imread(img_path, 0)
        pixel_values.extend(img.ravel())
        
        
        
# Plot the histogram
plt.hist(pixel_values, bins=256, range=[0, 256])
plt.title('Grayscale Histogram')
plt.xlabel('Grayscale Value')
plt.ylabel('Pixel Count')
plt.show()
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Path to the directories containing the X-ray images
covid_dir_path = "COVID-19 Dataset/X-ray/COVID/"
non_covid_dir_path = "COVID-19 Dataset/X-ray/Non-COVID/"

# Initialize lists to store pixel values of all images
covid_pixel_values = []
non_covid_pixel_values = []


# Loop through the first 100 images in the COVID directory in batches of 8 and append pixel values to the list
batch_size = 8
for i in range(0, 100, batch_size):
    batch = os.listdir(covid_dir_path)[i:i+batch_size]
    for filename in batch:
        img_path = os.path.join(covid_dir_path, filename)
        img = cv2.imread(img_path, 0)
        covid_pixel_values.extend(img.ravel())
        

# Loop through the first 100 images in the Non-COVID directory in batches of 8 and append pixel values to the list
for i in range(0, 100, batch_size):
    batch = os.listdir(non_covid_dir_path)[i:i+batch_size]
    for filename in batch:
        img_path = os.path.join(non_covid_dir_path, filename)
        img = cv2.imread(img_path, 0)
        non_covid_pixel_values.extend(img.ravel())
        
        
# Plot the histograms with different hues
plt.hist(covid_pixel_values, bins=256, range=[0, 256], alpha=0.5, color='red', label='COVID')
plt.hist(non_covid_pixel_values, bins=256, range=[0, 256], alpha=0.5, color='blue', label='Non-COVID')
plt.title('Grayscale Histogram')
plt.xlabel('Grayscale Value')
plt.ylabel('Pixel Count')
plt.legend()
plt.show()
```
